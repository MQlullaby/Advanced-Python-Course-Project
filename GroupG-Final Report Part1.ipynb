```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Coronavirus spread trend

```python
# Open coronavirus data file
filename = '2019_nCoV_data.csv'
df = pd.read_csv(filename,index_col=0)
df = df.loc[(df['Country']=='China')|(df['Country']=='Mainland China')]
df
```

```python
Confirmed = df["Confirmed"].groupby(df["Date"]).sum()
Deaths = df["Deaths"].groupby(df["Date"]).sum()
Recovered = df["Recovered"].groupby(df["Date"]).sum()
result = pd.concat([Confirmed,Deaths,Recovered],axis=1) 
result.index = pd.to_datetime(result.index)
result
```

```python
# Visualize the spread of covid-19 in China
result.plot()
```

```python
result.plot.line(subplots=True)
```

# Stock price change

```python
def read_file(filename):
    df = pd.read_csv(filename,index_col = 0,parse_dates = True)
    df['daily_return'] = df['Adj Close'].pct_change()
    return df
```

```python
# Entertainment
df_BILI = read_file("BILI.csv")
df_DOYU = read_file("DOYU.csv")
df_IQ = read_file("IQ.csv")
```

```python
adc_BILI = df_BILI['Adj Close']
adc_DOYU = df_DOYU['Adj Close']
adc_IQ = df_IQ['Adj Close']
plt.plot(adc_BILI, label = 'BILI')
plt.plot(adc_DOYU, label = 'DOYU')
plt.plot(adc_IQ, label = 'IQ')
plt.legend(loc='upper left')
plt.show()
```

```python
# E-commerce
df_BABA = read_file("BABA.csv")
df_JD = read_file("JD.csv")
df_MOGU = read_file("MOGU.csv")
df_PDD = read_file("PDD.csv")
df_SECO = read_file("SECO.csv")
df_YI = read_file("YI.csv")
df_YJ = read_file("YJ.csv")
```

```python
adc_BABA = df_BABA['Adj Close']
plt.plot(adc_BABA, label = 'BABA')
plt.legend(loc='upper left')
plt.show()
```

```python
adc_JD = df_JD['Adj Close']
adc_PDD = df_PDD['Adj Close']
plt.plot(adc_JD, label = 'JD')
plt.plot(adc_PDD, label = 'PDD')
plt.legend(loc='upper left')
plt.show()
```

```python
adc_MOGU = df_MOGU['Adj Close']
adc_SECO = df_SECO['Adj Close']
adc_YI = df_YI['Adj Close']
adc_YJ = df_YJ['Adj Close']
plt.plot(adc_MOGU, label = 'MOGU')
plt.plot(adc_SECO, label = 'SECO')
plt.plot(adc_YI, label = 'YI')
plt.plot(adc_YJ, label = 'YJ')
plt.legend(loc='upper left')
plt.show()
```

```python
# Logistics
df_BEST = read_file("BEST.csv")
df_ZTO = read_file("ZTO.csv")
```

```python
adc_BEST = df_BEST['Adj Close']
adc_ZTO = df_ZTO['Adj Close']
fig, ax1 = plt.subplots() 
adc_BEST.plot(ax=ax1,label = 'BEST') 
adc_ZTO.plot(secondary_y=True, ax=ax1, label = 'ZTO')
ax1.set_ylabel('BEST')
ax1.right_ax.set_ylabel('ZTO')
plt.legend(loc='upper left')
plt.show()
```

```python
# Tourism
df_TCOM = read_file("TCOM.csv")
df_TOUR = read_file("TOUR.csv")
```

```python
adc_TCOM = df_TCOM['Adj Close']
adc_TOUR = df_TOUR['Adj Close']
fig, ax1 = plt.subplots() 
adc_TCOM.plot(ax=ax1, label = 'TCOM') 
adc_TOUR.plot(secondary_y=True, ax=ax1, label = 'TOUR')
ax1.set_ylabel('TCOM')
ax1.right_ax.set_ylabel('TOUR')
plt.legend(loc='upper left')
plt.show()
```
