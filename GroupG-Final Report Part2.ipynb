# Regression analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
# Read files 
def read_file(filename):
    df = pd.read_csv(filename,index_col = 0,parse_dates = True)
    df['daily_return'] = df['Adj Close'].pct_change()
    return df
```

```python
df_SP500 = read_file("GSPC.csv")
# Entertainment
df_BILI = read_file("BILI.csv")
df_DOYU = read_file("DOYU.csv")
df_IQ = read_file("IQ.csv")
# E-commerce
df_BABA = read_file("BABA.csv")
df_JD = read_file("JD.csv")
df_MOGU = read_file("MOGU.csv")
df_PDD = read_file("PDD.csv")
df_SECO = read_file("SECO.csv")
df_YI = read_file("YI.csv")
df_YJ = read_file("YJ.csv")
# Logistics
df_BEST = read_file("BEST.csv")
df_ZTO = read_file("ZTO.csv")
# Tourism
df_TCOM = read_file("TCOM.csv")
df_TOUR = read_file("TOUR.csv")
```

```python
# Create a scatter plot to calculate daily returns
def scattor_plot(df_Stock,Stock):
    xret = df_SP500['daily_return'][1:]
    yret = df_Stock['daily_return'][1:]

    plt.title(Stock+ " vs SP500 - Daily Returns")
    plt.scatter(xret, yret)
    plt.xlabel('SP500 returns')
    plt.ylabel(Stock+ ' returns')
    plt.show()
    return
```

```python
# Create a linear regression model
def linear_regission(df_Stock):
    xret = df_SP500['daily_return'][1:]
    yret = df_Stock['daily_return'][1:]
    
    indvars = sm.add_constant(xret) #
    model = sm.OLS(yret, indvars).fit()
    summary = model.summary()
    print(summary)
    return
```

## Entertainment

```python
# BILI
df_Stock = df_BILI
Stock = 'BILI'
plot = scattor_plot(df_Stock,Stock)
lr = linear_regission(df_Stock)
```
